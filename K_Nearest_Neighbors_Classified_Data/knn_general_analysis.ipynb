```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
```

```python
df=pd.read_csv("classified_data.csv", index_col=0)
```

```python
df.head()
```

```python
# We will use this library to do some of the preprocessing to scale the data itself
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
# Now we want to fit this data to the feature data from the classified data set
scaler.fit(df.drop('TARGET CLASS', axis=1))
```

```python
# Now that we have fitted that data, we will need to transform that data
scaled_features=scaler.transform(df.drop('TARGET CLASS', axis=1))
```

```python
scaled_features
```

```python
# Now let us create a dataframe for the new scaled features
scaled_df=pd.DataFrame(data=scaled_features, columns=df.columns[0:len(df.columns)-1])
```

```python
scaled_df
```

```python
# Now let us perform the test train split
from sklearn.model_selection import train_test_split
```

```python
X=scaled_features
y=df['TARGET CLASS'].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=101)
```

```python
# Now lets import the k nearest neighbor stuff from sklearn
from sklearn.neighbors import KNeighborsClassifier as KNN
```

```python
knn = KNN(n_neighbors=3)
```

```python
# Now we have to fit the data to our model
knn.fit(X_train, y_train)
```

```python
# Now let's predict based on the test data set that we devised
predictions=knn.predict(X_test)
```

```python
print(predictions)
```

```python
# Now let us evaluate our model
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test, predictions))
print(confusion_matrix(y_test, predictions))
```

```python
# Let us now calculate the error rate of the nearest neighbors algorithm on a varying set of neighbors
error_rate=[]

for i in range(1,100):
    knn=KNN(n_neighbors=i)
    knn.fit(X_train, y_train)
    prediction=knn.predict(X_test)
    error_rate.append(np.mean(prediction!=y_test))

    
```

```python
# The following line of code will be used for the graphics
plt.figure(figsize=(20,20))
plt.plot(range(1,100), error_rate, color='blue', linestyle='dashed', marker='o', markerfacecolor='red', markersize=10)
plt.title('Error Rate v. K Value')
plt.xlabel('K-Value')
plt.ylabel('Error Rate')
```

```python
'''
From above we find that we can find the lowest error rate with a k-value of somewhere near 40
So let us test te model around this point
'''
knn = KNN(n_neighbors=40)
knn.fit(X_train, y_train)
predictions=knn.predict(X_test)

print(classification_report(y_test, predictions))
print(confusion_matrix(y_test, predictions))
```
