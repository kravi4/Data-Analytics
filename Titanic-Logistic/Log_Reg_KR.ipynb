```python
# Let us now programtically compute the average age from each class 
def compute_avg_age(class_num):
    Age_c_df=train_df[train_df['Pclass']==class_num][['Age']].copy()
    Age_c_df=Age_c_df.dropna()
    avg_c=mean(Age_c_df['Age'].values.tolist())
    tc_int_avg_c=int(avg_c)
    return tc_int_avg_c
    
avg_c1=compute_avg_age(1)
avg_c2=compute_avg_age(2)
avg_c3=compute_avg_age(3)
```

```python
# Now let us come up with a program to impute the null data in the age columns based on the class that each person came from
def impute_age(cols):
    age=cols[0]
    pclass=cols[1]
    
    if (pd.isnull(age)):
        if pclass==1:
            return avg_c1
        elif pclass==2:
            return avg_c2
        elif pclass==3:
            return avg_c2
    else:
        return age
```

```python
# Let us now apply the function to our training dataframe
train_df['Age']=train_df[['Age', 'Pclass']].apply(impute_age, axis=1)
sns.heatmap(train_df.isnull())

'''
Notice that in the heatmap that there is no null data for the age
'''
```

```python
# Since there is too much null data for the cabins, let us drop that column
train_df=train_df.drop('Cabin', axis=1)
```

```python
# When Visualizing again, notice that we have dropped the cabin data
sns.heatmap(train_df.isnull())
```

```python
# Now let's drop the last data point from the embarked column
train_df.dropna(inplace=True)
```

```python
sns.heatmap(train_df.isnull())
```

```python
# Now we have to come up with a way to deal with the categorical data such as the gender column and the embarked columns
sex=pd.get_dummies(train_df['Sex'], drop_first=True)
embarked=pd.get_dummies(train_df['Embarked'], drop_first=True)
```

```python
# Now let us concatenate the data frames altogether
# axis=1 specifies the columns
train_df=pd.concat([train_df, sex, embarked], axis=1)
```

```python
train_df.head()
```

```python
# Now let us drop the Embarked, Sex, Ticket, and the name columns
train_df=train_df.drop(['Sex', 'Embarked', 'Name', 'Ticket'], axis=1)
```

```python
train_df.head()
```

```python
# We also will probably want to drop the passenger id as well
train_df=train_df.drop("PassengerId", axis=1)
```

```python
train_df.head()
```

```python
# Note that in this instance the PClass is a categorical columns and we have to use a get dummies to seperate it as such
pclass=pd.get_dummies(train_df['Pclass'], drop_first=True)
pclass
train_df
```

```python
train_df=pd.concat([train_df, pclass], axis=1)
train_df
```

```python
train_df=train_df.drop("Age", axis=1)
```

```python
train_df.info()
```

```python
# Now lets set the x and y variables that we will feed into our logistic regression function
X=train_df.drop('Survived', axis=1)
y=train_df['Survived']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression as log
```

```python
# Creating an instance of the logistic regression model
logmodel=log()
```

```python
# Training the Logistic Regression model
logmodel.fit(X_train,y_train)
```

```python
# Getting the prediction values
predictions=logmodel.predict(X_test)
```

```python
# Now lets print the classification results
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,predictions))
```
